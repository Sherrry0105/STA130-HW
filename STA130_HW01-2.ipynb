```python
col_to_delete = 'Kuks', 'Quaas', 'Moans', 'Tail flags', 'Tail twitches'
```

```python
import pandas as pd

# Load the dataset
url = 'https://data.cityofnewyork.us/api/views/vfnx-vebw/rows.csv'
df = pd.read_csv(url)

# Before report: Check shape and count missing values in each column
print("Before cleanup:")
print("Shape of dataset:", df.shape)
print("Missing values in each column:\n", df.isnull().sum())
```

```python
# Step 1: Delete columns that are not useful or have too many missing values
# Assume 'col_to_delete' has too many NaN values and is irrelevant

# Set a threshold for the percentage of missing values (e.g., 50%)
threshold = 0.5

# Identify columns where the percentage of missing values exceeds the threshold
cols_to_delete = df.columns[df.isnull().mean() > threshold]

# Delete those columns
df.drop(cols_to_delete, axis=1, inplace=True)

# Print the columns that were deleted
print("Deleted columns with too many missing values:", cols_to_delete)

# Check the new shape of the dataset after deletion
print("New shape of the dataset:", df.shape)
```

```python
# Step 2: Drop rows with any remaining missing values
df_cleaned = df.dropna()
```

```python
# After report: Check shape and missing values again
print("\nAfter cleanup:")
print("Shape of cleaned dataset:", df_cleaned.shape)
print("Missing values in each column:\n", df_cleaned.isnull().sum())
```
